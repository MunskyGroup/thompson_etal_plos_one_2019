After performing feature selection using the training data, we may wish to know how the model performs with testing data using the reduced feature sets. The default choice for reduced features is the joint set of selected features between the three methods as described in the paper. In this tutorial, we can examine how well the model performs with test data using the features selected by each method to produce figures similar to Fig 2C. 

Import ML_microbiome, pandas, numpy 

```python
import pandas as pd
import numpy as np
# import RFINN
from RFINN import *
```

Import the training and testing data and generate the feature selection table 

```python
# import training and test data as pandas dataframes
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')
targets = pd.read_csv('DOC_targets.csv')
```

Instantiate a feature selection object with the data and determine the default feature selection table that includes the consensus set of selected features. 

```python
FS = FeatureSelection(train_data, test_data, targets)
FS_table = pd.read_csv('FS_table.csv')
```

Plot Random Forest model performance as a function of included features from the consensus set and from random forest feature ranking alone. This is done using the CompareFeatureSelection() function which compares performance of a specified model using the consensus set of features to the feature ranking determined by a specified feature ranking method.  

```python
FS.CompareFeatureSelection(FS_table=FS_table, model='Random Forest', compare_method='Random Forest', min_ftrs=5, max_ftrs=100, bin_size=5, fname='RF_compareFS.png')
```

The random forest model does better using the features ranked exclusively by random forest. 

```python
FS.CompareFeatureSelection(FS_table=FS_table, model='Neural Network', compare_method='Neural Network', min_ftrs=5, max_ftrs=100, bin_size=5, fname='NN_compareFS.png')
```

Once features exceed roughly 30, the neural network model performs better using neural network features compared to the consensus set. 

One would expect that individual models perform best with the features that were ranked by that model. Because we do not know 'ground truth', the consensus set of features provide the most confident set of taxa that were considered important by all methods (indicator species, random forest, and neural network).
