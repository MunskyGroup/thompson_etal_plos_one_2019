# Run Random Forest and Neural Network Prediction Model with Hyper-Parameter Tuning

Data must be uploaded as Pandas DataFrames with each sample in rows and each feature in columns. The data file with targets is slightly different, with sample IDs as columns. 

This is accomplished by first importing the Pandas library: 

```python
import pandas as pd
# numpy imported for later use 
import numpy as np
```

Upload DataFrames of training, testing, and target data using pd.read_csv()

```python
# import training and test data as pandas dataframes
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')
targets = pd.read_csv('DOC_targets.csv')
```

Import the Model class from ML_microbiome: 

```python
from ML_microbiome import Model
```

Let's start by setting up a Random Forest Regression model: 

```python
# instantiate model class
model = Model('Random Forest', train_data, test_data, targets)
```

Training and Testing the model followed by hyper-parameter optimization is accomplished with the following commands: 

```python
# optimize hyper-parameters 
print(model.tune_hyper_params())

# train model
model.train_model()

# test model
Y_test, Y_pred_test, R = model.test_model()
```

```python
# print prediction performance results 
print("Pearson's R: {:.3f}".format(R))
MSE = np.mean((Y_test - Y_pred_test)**2)
print("Mean Squared Error: {:.3f}".format(MSE))
```

Implementing a Neural Network model requires specifying 'Neural Network' when instantiating the model: 

```python
# instantiate model class as Neural Network
model = Model('Neural Network', train_data, test_data, targets)
```

Training and Testing the model is the same as before: 

```python
# tune hyper parameters for model
print(model.tune_hyper_params())

# train model
model.train_model()

# test model
Y_test, Y_pred_test, R = model.test_model()
```

```python
# print prediction performance results 
print("Pearson's R: {:.3f}".format(R))
MSE = np.mean((Y_test - Y_pred_test)**2)
print("Mean Squared Error: {:.3f}".format(MSE))
```
